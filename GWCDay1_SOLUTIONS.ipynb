### **Integers**

To define an integer, use the following syntax:



**Note:** "myint" is just the name of a variable that is holding the integer value 7.

```
myint = 7
print(myint)
```

### **Floats**

To define a floating point number, you may use one of the following notations:

```
myfloat = 7.0
print(myfloat)
myfloat = float(7)
print(myfloat)
```

### **Strings**

Strings can be defined either with a single quote or double quotes. 

```
#single quote
mystring = 'hello'
print(mystring)

#double quote
mystring = "hello"
print(mystring)
```

**Why even bother?**

The difference between the two is that using double quotes makes it easy to include apostrophes (whereas these would end the string if using single quotes).

**Here is an example...**

```
#using double quotes - nice
mystring = "Don't worry about apostrophes."
print(mystring)
```

## **Basic Operatiors**

Just like in math - there are 4 main operations you can do in Python (with integers & floats):

*   Addition (+)
*   Subtraction (-)
*   Multiplication (*)
*   Division (/)

Try to predict what the answer will be. Does Python follow order of operations?

```
number = 1 + 2 * 3 / 4.0
print(number)
```

You can edit the block above to calculate whatever you want. 

Another operator available is the modulo (%) operator, which returns the integer remainder of any division. 

**dividend % divisor = remainder**

```
remainder = 11 % 3
print(remainder)
```

Using two multiplication symbols makes a power relationship. (something to the power of something)

```
squared = 7 ** 2
cubed = 2 ** 3
print(squared)
print(cubed)
```

If you use a double slash (//) instead of the normal single slash (/) for division, then the answer given will also be an integer (even if there is a remainder). This is called **integer division**.

```
normaldiv = 5/2
print("This is normal division.")
print(normaldiv)

print("This is integer division.")
integerdiv = 5//2
print(integerdiv)
```

### **Using Operators with Strings**

Python supports concatenating *(putting together)* strings using the addition operator:

```
helloworld = "hello" + " " + "world"
print(helloworld)
```

Python also supports multiplying strings to form a string with a repeating sequence:



```
lotsofhellos = "hello\n" * 10
print(lotsofhellos)
```
